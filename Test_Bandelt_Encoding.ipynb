```python
import dendropy
from dendropy.calculate import treecompare
import os
```

```python
folder_list = ['Bergsten_2013', 'Broughton_2013', 'Brown_2012', 'Cannon_2016_dna']
# folder_list = ['Bergsten_2013', 'Cannon_2016_dna', 'Leache_2015', 'Prebus_2017', 'Smith_2014', 'Broughton_2013', 'Crawford_2012', 'McCormack_2013', 'Ran_2018_dna', 'Wu_2018_dna', 'Brown_2012', 'Faircloth_2013', 'Meiklejohn_2016', 'Richart_2015']
```

```python
training_dataset_prefix = "./Bandelt_Test_Data/"
# training_dataset_prefix = "training_big"
```

```python
for folder in folder_list:
    file_path = os.path.join(training_dataset_prefix, folder)
    file_path_write = os.path.join(training_dataset_prefix, 'tree_decoding_output', folder)
    files_list = os.listdir(file_path)
    for file in files_list:
        if file.find('.nex.treefile') > 0:
            file_base_name = file[0:file.find('.nex.treefile')]
            
            taxon_namespace = dendropy.TaxonSet()
            tree_original = dendropy.Tree.get(path=os.path.join(file_path, file_base_name+'.nex.treefile'), schema="newick", taxon_set=taxon_namespace)
            
            bandelt_encode_list, encode_dict, decode_dict = dendropy.calculate.treemeasure.Bandelt_encode(tree_original)
            encode_newick_string = dendropy.calculate.treemeasure.Bandelt_decode(bandelt_encode_list, decode_dict)
            
            tree_decode = dendropy.Tree.get(data=encode_newick_string, schema="newick", taxon_set=taxon_namespace)
            print("** File: ", os.path.join(file_base_name+'.nex.treefile'), "   Original Tree and decoded tree distance: ", treecompare.symmetric_difference(tree_original, tree_decode))
            
```
